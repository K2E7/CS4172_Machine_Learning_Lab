```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd

# Split your data into features (X) and the target variable (y)
X = df.drop(columns=["Cover_Type"])  # Adjust the column name if needed
y = df["Cover_Type"]

# Create a Random Forest classifier
rf = RandomForestClassifier(n_estimators=100, random_state=0)  # You can adjust hyperparameters as needed

# Fit the model to your data
rf.fit(X, y)

# Get feature importances
feature_importances = rf.feature_importances_

# Create a DataFrame to display feature names and their importance scores
importance_df = pd.DataFrame({"Feature": X.columns, "Importance": feature_importances})

# Sort the DataFrame by importance score in descending order
importance_df = importance_df.sort_values(by="Importance", ascending=False)

# Print the sorted feature importance
print(importance_df)
```

```python
import numpy as np

def get_low_importance_columns(threshold=0,importance_df=importance_df):
    """
    Get columns with importance scores below the specified threshold.

    Parameters:
        importance_df (pd.DataFrame): DataFrame with feature names and importance scores.
        threshold (float): Importance threshold.

    Returns:
        list: List of column names with importance scores below the threshold.
    """
    low_importance_columns = importance_df[importance_df['Importance'] < threshold]['Feature'].tolist()
    return low_importance_columns
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
# Standardize features (important for SVM)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Train One-vs-Rest (OvR) SVM
ovr_svm = SVC(kernel="linear", decision_function_shape="ovr", random_state=42)
ovr_svm.fit(X_train, y_train)

# Train One-vs-One (OvO) SVM
ovo_svm = SVC(kernel="linear", decision_function_shape="ovo", random_state=42)
ovo_svm.fit(X_train, y_train)

# Make predictions
y_pred_ovr = ovr_svm.predict(X_test)
y_pred_ovo = ovo_svm.predict(X_test)

# Calculate accuracy for OvR and OvO
accuracy_ovr = accuracy_score(y_test, y_pred_ovr)
accuracy_ovo = accuracy_score(y_test, y_pred_ovo)

# Create confusion matrices
cm_ovr = confusion_matrix(y_test, y_pred_ovr)
cm_ovo = confusion_matrix(y_test, y_pred_ovo)

# Print the model parameters
print("OvR SVM Parameters:")
print(ovr_svm)
print("\nOvO SVM Parameters:")
print(ovo_svm)

# Print the difference between OvR and OvO
print(f"\nDifference between OvR and OvO:\n")
print(f"OvR SVM has {len(ovr_svm.classes_)} binary classifiers, one for each class.")
print(f"OvO SVM has {len(ovo_svm.classes_)} * ({len(ovo_svm.classes_) - 1}) / 2 binary classifiers.")
```

```python
def SVM_Comp(t=0, df=df):
  imp_t = get_low_importance_columns(threshold=t)
  print(imp_t)
  df_filtered = df.drop(columns=imp_t)
  X = df_filtered.drop('Cover_Type', axis = 1)
  y = df_filtered['Cover_Type']
  undersample = RandomUnderSampler(sampling_strategy='not minority')
  X, y = undersample.fit_resample(X, y)


  X_train, _X, y_train, _y = train_test_split(X, y, train_size=0.8)
  X_test, X_val, y_test, y_val = train_test_split(_X, _y, train_size = 0.5)

  # Standardize features (important for SVM)
  scaler = StandardScaler()
  X_train = scaler.fit_transform(X_train)
  X_test = scaler.transform(X_test)

  # Train One-vs-Rest (OvR) SVM
  ovr_svm = SVC(kernel="linear", decision_function_shape="ovr", random_state=42)
  ovr_svm.fit(X_train, y_train)

  # Train One-vs-One (OvO) SVM
  ovo_svm = SVC(kernel="linear", decision_function_shape="ovo", random_state=42)
  ovo_svm.fit(X_train, y_train)

  # Make predictions
  y_pred_ovr = ovr_svm.predict(X_test)
  y_pred_ovo = ovo_svm.predict(X_test)

  # Calculate accuracy for OvR and OvO
  accuracy_ovr = accuracy_score(y_test, y_pred_ovr)
  accuracy_ovo = accuracy_score(y_test, y_pred_ovo)

  # Create confusion matrices
  cm_ovr = confusion_matrix(y_test, y_pred_ovr)
  cm_ovo = confusion_matrix(y_test, y_pred_ovo)

  # Print the model parameters
  print("OvR SVM Parameters:")
  print(ovr_svm)
  print("\nOvO SVM Parameters:")
  print(ovo_svm)

  # Print the difference between OvR and OvO
  print(f"\nDifference between OvR and OvO:\n")
  print(f"OvR SVM has {len(ovr_svm.classes_)} binary classifiers, one for each class.")
  # Calculate the number of binary classifiers in OvO SVM
  num_classes = len(ovo_svm.classes_)
  num_binary_classifiers = (num_classes * (num_classes - 1)) // 2

  # Print the number of binary classifiers
  print(f"OvO SVM has {num_binary_classifiers} binary classifiers.")

  # Plot confusion matrices
  plt.figure(figsize=(12, 5))

  plt.subplot(1, 2, 1)
  sns.heatmap(cm_ovr, annot=True, fmt="d", cmap="Blues", xticklabels=np.unique(y), yticklabels=np.unique(y))
  plt.title(f"OvR SVM\nAccuracy: {accuracy_ovr:.2f}")
  plt.xlabel("Predicted")
  plt.ylabel("True")

  plt.subplot(1, 2, 2)
  sns.heatmap(cm_ovo, annot=True, fmt="d", cmap="Blues", xticklabels=np.unique(y), yticklabels=np.unique(y))
  plt.title(f"OvO SVM\nAccuracy: {accuracy_ovo:.2f}")
  plt.xlabel("Predicted")
  plt.ylabel("True")

  plt.tight_layout()
  plt.show()
```
